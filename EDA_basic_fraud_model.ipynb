```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('C:/Users/Acer/Desktop/fraud_model/card_data.csv') 
print(data.columns)
data.head()
```

```python
print(data.shape)
print(data.describe())
```

```python
import seaborn as sns
amount = [data['Amount'].values]
sns.distplot(amount) #plotting distribution
```

```python
time = data['Time'].values #no. of seconds elapsed from the first transaction and this transaction in DB
sns.distplot(time) #plotting distribution 
```

```python
from matplotlib import gridspec
# distribution of anomalous features
features = data.iloc[:,0:28].columns

plt.figure(figsize=(12,28*4))
gs = gridspec.GridSpec(28, 1)
for i, c in enumerate(data[features]): #loop through all features
    ax = plt.subplot(gs[i])
    sns.distplot(data[c][data.Class == 1], bins=30) #label = fraud
    sns.distplot(data[c][data.Class == 0], bins=30) #label = non-fraud
    ax.set_xlabel('')
    ax.set_title('histogram of feature: ' + str(c))
plt.show()
```

```python
# Plot histograms of each parameter 
data.hist(figsize = (20, 20))
import matplotlib.pyplot as plt
plt.show()
```

```python
# Segregate fraud & non-fraud cases in dataset

fraud = data[data['Class'] == 1]
valid = data[data['Class'] == 0]
#print(fraud)
#print(valid)


print('Fraud Cases:', len(data[data['Class'] == 1]))
print('Valid Transactions:', len(data[data['Class'] == 0]))

fraud_fraction = len(fraud)/float(len(valid))
print('Fraction of Fraud Cases:', fraud_fraction)
```
